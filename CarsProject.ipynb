## **Data Collection**

```python
import pandas as pd

df = pd.read_csv('cars.csv')

print("Yay! The data is loaded.")
```

```python
print("Here are the very first 5 cars in our list:")
display(df.head())
```

```python
print("\nHere are the very last 5 cars in our list:")
display(df.tail())
```

```python
mini_table = df[['Car Name', 'Color', 'Energy', 'Price-$']]
print("\nHere is a mini-table with just a few specific columns:")
display(mini_table.head())
```

## **Data Cleaning**

```python
print(f"Before cleaning, cars count {len(df)} ")
```

```python
df = df.drop_duplicates()
```

```python
df['Feedback'] = df['Feedback'].fillna('No Feedback') #to remove empty feedbacks
```

```python
df['Sold Date'] = df['Sold Date'].fillna('Not Sold Yet') #to clean sold/not sold column
```

```python
df['Sold Price-$'] = df['Sold Price-$'].fillna(0) #if its not sold
```

```python
df = df.dropna(subset=['Car Name', 'Price-$']) #missing ones
```

## **Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", font_scale=1.2)

# Chart 1: Bar Chart for Car Types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Car Type', palette='Set2')
plt.title('1. Most Popular Car Types', fontweight='bold')
plt.show()
```

```python
# Chart 2: Bar Chart for Fuel
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Energy', palette='pastel')
plt.title('2. What Fuel Do the Cars Use?', fontweight='bold')
plt.show()
```

```python
# Chart 3: Line Chart for Price vs Year
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Manufactured Year', y='Price-$', color='green', linewidth=3)
plt.title('3. Car Price by Year Made', fontweight='bold')
plt.show()
```

```python
# Chart 4: Boxplot for Price Ranges
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Car Type', y='Price-$', palette='Set3')
plt.title('4. Price Range for Each Car Type', fontweight='bold')
plt.show()
```

```python
# Chart 5: Sideways Bar Chart for Top Agents
plt.figure(figsize=(10, 6))
top_agents = df['Sales Agent Name'].value_counts().index[:5]
sns.countplot(data=df, y='Sales Agent Name', palette='viridis', order=top_agents)
plt.title('5. Top 5 Best Selling Sales Agents', fontweight='bold')
plt.show()
```

```python
# Chart 6: Average Price by Gearbox
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Gearbox', y='Price-$', palette='magma')
plt.title('6. Average Price based on Gearbox (Manual vs Auto)', fontweight='bold')
plt.show()
```

```python
# Chart 7: Histogram for Mileage
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Mileage-KM', bins=30, color='purple', kde=True)
plt.title('7. How Much Have These Cars Been Driven?', fontweight='bold')
plt.show()
```

```python
# Chart 8: Heatmap for Secret Connections
plt.figure(figsize=(10, 8))
# We pick a few specific numbers to compare
numeric_columns = df[['Price-$', 'Mileage-KM', 'Engine Power-HP', 'Manufactured Year', 'Margin-%']]
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('8. Secret Connections Map', fontweight='bold')
plt.show()
```

```python
# Chart 9: The Donut Chart
plt.figure(figsize=(8, 8))
status_counts = df['Car Sale Status'].value_counts()

plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], startangle=90)

center_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title('9. Donut Chart: Are we selling our cars?', fontweight='bold')
plt.show()
```

```python
# Chart 10: Double Bar Chart
plt.figure(figsize=(12, 6))
# 'hue' is the magic word that splits the bars side-by-side
sns.barplot(data=df, x='Car Type', y='Price-$', hue='Gearbox', palette='Set1')
plt.title('10. Double Bar: Automatic vs Manual Prices for Each Car Type', fontweight='bold')
plt.show()
```
